```python
from operator import itemgetter
genre_list = ["Horror","Romance","Comedy","History","Adventure","Action"]
print("Please Join to this watching movie survey and let us know your 3 favorite genre from the list below :")
print("\n".join(genre_list))
print()
survey = {}
while True:
    number_of_person  = input("Please Enter the number of person :")
    if number_of_person.isdigit():
        if int(number_of_person)>0 :
            break

for i in range(int(number_of_person)):
    name = input("Enter the Name :")
    fav_list = []
    for j in range(3):
        while True:
            favorites = input("Enter your {0} favaorite genre :".format(j+1))
            if favorites.casefold().capitalize() in genre_list and favorites.casefold() not in fav_list:
                fav_list.append(favorites.casefold())
                break
            else :
                print("Just from the list and Not Duplicate !")
    survey[name] = fav_list     

# survey = {"amir":["horror","action","comedy"],
#           "hassan":["adventure","action","romance"]}    

# Get out Survey value into a new list as a raw data
raw_list =[]
group_list= list(survey.values())
for lst in group_list:
    for items in lst:
        raw_list.append(items)
#print("raw list : {}".format(raw_list))

# Count number of each genre & save in dictionary
final_list = []
for i in range(len(genre_list)):
    short_list = []
    #final_dic[genre_list[i]]=raw_list.count(genre_list[i].casefold())
    short_list.append(str(raw_list.count(genre_list[i].casefold())))
    short_list.append(genre_list[i])
    final_list.append(short_list)
#print("final list without sorting : {}".format(final_list))

# sorted_list_temp = sorted(final_list,reverse=True)
# print("sortedtemp : {}".format(sorted_list_temp))

sorted_list = sorted(final_list,reverse=True)
print("sorted : {}".format(sorted_list))
for i in range(len(sorted_list)):
    j=0
    print("{} : {}".format(sorted_list[i][j+1],sorted_list[i][j]))
print("------- sorry couldn't sort by alphabet !! ---------")

# #-------------test not completed
# mylist = [[2, 'Horror'], [2, 'Comedy'], [2, 'Action'], [1, 'Romance'], [1, 'History'], [1, 'Adventure']]
# #-----------convert to Dic
# test = {2:"horro comdey action",
#         1:"romance  adventure",
#         0:"history"}
# print("-----------")

# listt = (list((test.values())))
# listn = list(test.keys())
# print("listt : {}".format(listt))
# print("listn: {}".format(listn))
# print("-----------")
# list_t =[]
# for i in range(len(listt)):
#     list_t.append(listt[i].split())
# print(list_t)
# print(listn)
# print("-----------")
# list_f=[]
# for i in range(len(listn)):
#     list_f.append(sorted(list_t[i]))
# print(list_f)
# print(listn)
# for i in range(len(listn)):
#     for j in range(len(list_f[i])):
#         print("{} : {}".format(list_f[i][j],listn[i]))

    
    

```
