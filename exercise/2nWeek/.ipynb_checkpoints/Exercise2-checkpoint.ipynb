```python
import operator
import re
iran = []
spain = []
morocco = []
portugal = []
temp = []
print("Please Enter game result as below(first Num to the left team):")
#-----------------------------------------Getting Scores
#-------------------------------Iran-spain
#------------Get Scores
while True:
    iran_spain = input ("Iran-Spain : ") 
    goals = re.findall(r"[\w']+", iran_spain)
    if len(goals) == 2 :
         if goals[0].isdigit() and goals[1].isdigit():
            break
#------Set for each Team
goals.append(int(goals[0])-int(goals[1]))
iran.append(goals)
temp.append(goals[1])
temp.append(goals[0])
temp.append(int(temp[0])-int(temp[1]))
spain.append(temp)
# print("Iran : {}".format(iran))
# print("Spain : {}".format(spain))
#------------------------------Iran-Portugal
#------------Get Scores
temp=[]
while True:
    iran_portugal = input ("Iran-Portugal : ") 
    goals = re.findall(r"[\w']+", iran_portugal)
    if len(goals) == 2 :
         if goals[0].isdigit() and goals[1].isdigit():
            break
#------Set for each Team
goals.append(int(goals[0])-int(goals[1]))
iran.append(goals)
temp.append(goals[1])
temp.append(goals[0])
temp.append(int(temp[0])-int(temp[1]))
portugal.append(temp)
# print("Iran : {}".format(iran))
# print("Portugal : {}".format(portugal))
#------------------------------Iran-Morocco
#------------Get Scores
temp=[]
while True:
    iran_morocco = input ("Iran-Morocco : ")
    goals = re.findall(r"[\w']+", iran_morocco)
    if len(goals) == 2 :
         if goals[0].isdigit() and goals[1].isdigit():
            break
#------Set for each Team
goals.append(int(goals[0])-int(goals[1]))
iran.append(goals)
temp.append(goals[1])
temp.append(goals[0])
temp.append(int(temp[0])-int(temp[1]))
morocco.append(temp)
# print("Iran : {}".format(iran))
# print("Morocco : {}".format(morocco))
#------------------------------Spain-Portugal
#------------Get Scores
temp=[]
while True:
    spain_portugal = input ("Spain-Portuga : ") 
    goals = re.findall(r"[\w']+", spain_portugal)
    if len(goals) == 2 :
         if goals[0].isdigit() and goals[1].isdigit():
            break
#------Set for each Team
goals.append(int(goals[0])-int(goals[1]))
spain.append(goals)
temp.append(goals[1])
temp.append(goals[0])
temp.append(int(temp[0])-int(temp[1]))
portugal.append(temp)
# print("Spain : {}".format(spain))
# print("Portugal : {}".format(portugal))
#------------------------------Spain-Morocco
#------------Get Scores
temp=[]
while True:
    spain_morocco = input ("Spain-Morocco : ") 
    goals = re.findall(r"[\w']+", spain_morocco)
    if len(goals) == 2 :
         if goals[0].isdigit() and goals[1].isdigit():
            break
#------Set for each Team
goals.append(int(goals[0])-int(goals[1]))
spain.append(goals)
temp.append(goals[1])
temp.append(goals[0])
temp.append(int(temp[0])-int(temp[1]))
morocco.append(temp)
# print("Spain : {}".format(spain))
# print("Morocco : {}".format(morocco))
#------------------------------Portugal-Morocco
#------------Get Scores
temp=[]
while True:
    portugal_morocco = input ("Portugal-Morocco : ") 
    goals = re.findall(r"[\w']+", portugal_morocco)
    if len(goals) == 2 :
         if goals[0].isdigit() and goals[1].isdigit():
            break
#------Set for each Team
goals.append(int(goals[0])-int(goals[1]))
portugal.append(goals)
temp.append(goals[1])
temp.append(goals[0])
temp.append(int(temp[0])-int(temp[1]))
morocco.append(temp)
# print("Portugal : {}".format(portugal))
# print("Morocco : {}".format(morocco))
#--------[Zade , khorde , Tafazol]-----------------------------------------
print("*" * 30,"[ Zade Khorde Tafazol ]","*" * 30)
print("Iran : {}".format(iran))
print("Spain : {}".format(spain))
print("Portugal : {}".format(portugal))
print("Morocco : {}".format(morocco))
print("*" * 60)
# #-------------Show The Result------------------------------------

# iran = [['2', '2', 0], ['2', '1', 1], ['1', '2', -1]]
# spain = [['2', '2', 0], ['2', '2', 0], ['3', '1', 2]]
# portugal = [['1', '2', -1], ['2', '2', 0], ['2', '1', 1]]
# morocco = [['2', '1', 1], ['1', '3', -2], ['1', '2', -1]]

teams ={"iran":iran,
        "spain":spain,
        "portugal":portugal,
        "morocco":morocco
       }
final_res = []
    
for key in teams:
    draw =0
    lose = 0
    win = 0
    goal_difference = 0
    point = 0
    per_country = []
    each_play = teams.get(key)    
    for i in range(3) :
        if each_play[i][2]==0:
            draw += 1
            point +=1
        elif each_play[i][2]<0 :
            lose +=1
        else :
            win+=1
            point +=3
        goal_difference +=each_play[i][2] 
    per_country.append(point)
    per_country.append(win)
    per_country.append(key)
    per_country.append(draw)
    per_country.append(lose)
    per_country.append(goal_difference)
    final_res.append(per_country)
#-----------------Sorting----------------------------------------
list1 = sorted(final_res, key=operator.itemgetter(2))
list2 = sorted(list1, key = operator.itemgetter(0,1), reverse=True)
for items in list2 :
        print("{}  wins : {} , loses : {} ,  draws : {} , Goal Difference : {} , Point : {}"
              .format(items[2].capitalize(),items[1],items[4],items[3],items[5],items[0]))


    
```
