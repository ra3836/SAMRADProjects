```python
class Pagination(object):
        
    def __init__(self,items=[],pagesize=10):
        self.items = items
        self.pagesize = pagesize
        self.main_list = []
        self.sub_list = []
        self.index = 0
        
        self.set_pages()
        

        
    def set_pages(self):
        count = 1
        for item in self.items:
            if count == self.pagesize:
                self.sub_list.append(item)
                self.main_list.append(self.sub_list)
                self.sub_list = []
                count = 1
            else:
                self.sub_list.append(item)
                count += 1
        if count > 0:
            self.main_list.append(self.sub_list)
        return self.main_list

        
        
    def getvisibleitems(self):
        return self.main_list[self.index]
    
    
    def nextpage(self):
        if self.index == len(self.main_list)-1 :
            print("No More Page !")
        else:    
            self.index +=1 
            return self

    
    def previuspage(self):
        if self.index == 0 :
            print("Here is the first Page !")
        else :
            self.index -=1
            return self
    
    def lastpage(self):
        self.index = len(self.main_list)-1
        return self
    
    def firstpage(self):
        self.index = 0
        return self

    def gotopage(self,num):
        if num not in range(len(self.main_list)-1):
            print("Index Out Of Range !")
        else :
            self.index = num
            return self
    
    
alphabetlist = "a b c d e f g h i j k l m n o p q r s t u v w x y z".split(" ")
print(alphabetlist)
p = Pagination(alphabetlist,4)

print(p.getvisibleitems())

p.nextpage()
print(p.getvisibleitems())

p.nextpage()
print(p.getvisibleitems())

p.previuspage()
print(p.getvisibleitems())

p.lastpage()
print(p.getvisibleitems())

p.gotopage(5)
print(p.getvisibleitems())

p.nextpage().nextpage()
print(p.getvisibleitems())




```
